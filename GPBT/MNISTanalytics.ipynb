```python
import pandas as pd
from numpy import genfromtxt
import numpy as np
from scipy.interpolate import *
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import time
import math
from scipy.signal import savgol_filter
```

```python
def fonc(data):
    itera = data[:,5:6]
    itera= itera[~np.isnan(itera)]
    dat = data[:,0:2]
    dat = dat[~np.isnan(dat)]
    dat = dat.reshape(-1,2)
    dat = dat[np.argsort(itera,axis=-1)]
    return dat
```

```python
def maxof(a):
    ma = 0;
    ta = 0;
    for i in range(250):
        if(a.shape[0]<=i):
            a=np.concatenate((a,(np.array([np.array([ma,ta])]))))
            print(a.shape)
        else:
            if(ma<a[i,0]):
                ta = a[i,1]
            ma = max(ma,a[i,0]) 
            a[i,0] = ma
            a[i,1] = ta
    return a[:250] 
```

```python
def iteration_corector(liste, num_config):
    for i in range(len(liste[:])):
        liste[i] = math.floor(i/num_config)
```

```python
def getall(a):
    f = [b for b in a[:,1::2].mean(1)]
    g = [b for b in a[:,0::2].mean(1)]
    def h(x):
        return savgol_filter(np.ravel(x),3, 1)
    return (h(f), h(a[:,1::2].std(1)/2),
            h(g), h(a[:,0::2].std(1)/2))
```
