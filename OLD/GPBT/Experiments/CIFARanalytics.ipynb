```python
def maxof(a,n=51):
    ma = 0;
    ta = 0;
    for i in range(len(a[:,0])):
        if(a.shape[0]<=i):
            a=np.concatenate((a,(np.array([np.array([ma,ta])]))))
            print(a.shape)
        else:
            if(ma<a[i,0]):
                ta = a[i,1]
            ma = max(ma,a[i,0]) 
            a[i,0] = ma
            a[i,1] = ta
    return [savgol_filter(np.ravel(a[:,:1]),n, 1),
        savgol_filter(np.ravel(a[:,1:]),n, 1)]

#data -> pb2 -> last (bhob)
```
